```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set up random seed function 
np.random.seed(613)
# Generate random data to plot
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
# ax.scatter(x,y)
# ax.bar(x,y)
# ax.barh(x,y)
# ax.plot(x,y)
# create a histogram
ax.hist(y)
```

```python
# Set up fonts
font1 = {'family':'sans-serif', 'color':'blue', 'size': 20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

# create a line plot with title and labels using the tight layout
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
# Plot scatterplot using indigo star markers 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y, marker='*', color='indigo')
```

```python
# Plot scatterplot using linestyle and indigo star markers 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y, marker='*', color='indigo', linestyle='--', linewidth=2)
```

```python
# create a scatterplot with coloured markers and lines,
# showing a dotted grid
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y, 
        marker='*',
        markersize=12, 
        color='#7425b9', 
        linestyle='--', 
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')
ax.grid(linestyle='dotted', color='blue', linewidth='2')
```
