```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import scipy
import PIL 
import requests

# Set random seed for reproducibility
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# Plot line plot with both of our y variables on the same axes 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc = 'lower right',    
    frameon = True, # add frame around the legend
    fontsize = 12, # change the font size 
    ncol = 2, # specify number of columns 
    shadow = True)
fig.show()
```

```python
# Move legend to upper left corner to prevent obstruction of the plot.
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax. legend(loc = "upper left",
           bbox_to_anchor =(1, 1))
```

```python
# Add text to this plot by specifying the coordinates
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95, "This value is important!",
        ha = 'center', # aligment of text
        color = 'red', # modify font colour 
        size = 20) # modify font size 
```

```python
# Create an array of second set of data 
# np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np. array([110,180,240,99,220])
```

```python
# Outline number of columns, number of rows, figure size
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7, 3))

font1 = {'family': 'serif', 'color': 'indigo'}
font2 = {'family': 'serif', 'color': 'hotpink'}

# Separate (x,y) into two separate plots 
ax1.scatter(x1,y1, marker = 'o', color = 'red')
ax2.bar(x2, y2, color = 'green') 

# Set title and labels for x-axis
ax1.set_title('Important Scatterplot', fontdict= font2)
ax1.set_ylabel('Measure', fontdict =font1)
ax1.set_xlabel('Value', fontdict=font1)

# Annotate message on axis 1 and 2
ax1.annotate("This is important!", xy = (10,95), 
            xytext = (20, 94), arrowprops = dict(facecolor = 'hotpink',
                                                 arrowstyle = 'fancy'))

ax2.annotate("This is not important!", xy = (10,95),
             xytext = (20, 94), arrowprops = dict(facecolor = 'hotpink',
                                                 arrowstyle = 'fancy'))
```

```python
# Create mosaic for the subplots and use constrained format to fit all the subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = 'constrained')

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)  

# Add labels to the subplots 
someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)             
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18) 
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18) 
```

```python
# Create sample data
x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

# Define our figure and axes
fig, ax = plt.subplots(figsize=(7,3))

# Call both bar and plot elements to the same axes (ax)
ax.bar(x,y1, color = 'indigo')
ax.plot(x,y2, color = 'red')
```

```python
# Calculate standard deviation 
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = 'red')

# Plot standard deviation error bars
ax.errorbar(x,y2, yerr = y2_sd, fmt = "none",
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2) # set intervals 
```

```python
from PIL import Image # to open images 
import requests # to get images from URLs 
from io import BytesIO

# Get image from the internet 
response = requests.get('https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = 'red')

# Overlay a new axis on the figure to act as a container for the image
ax_image = fig.add_axes([0.1,0.11, 0.15, 0.35]) 

ax_image.imshow(image)
ax_image.axis('off')

# Save the visualization in a folder
path = 'C:/Users/samka/Documents/GitHub/visualization/02_activities/assignments/participation/'
filename = 'fig1a.png'
plt.savefig(path + filename, dpi=300)
```
